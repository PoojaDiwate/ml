```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/home/cslogin/mscii81/dataset/SalaryData.csv')
df
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(df[['Salary']])
```

```python
x_poly
```

```python
from sklearn.linear_model import LinearRegression
poly_reg = LinearRegression()
poly_reg.fit(x_poly,df.Salary)
```

```python
Y_pprice=poly_reg.predict(x_poly)
Y_pprice
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.scatter(df.Experience,df.Salary,color='red',marker='*')
plt.plot(df.Experience,poly_reg.predict(x_poly),color='blue')
```
