```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report 
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
data=pd.read_csv('/home/cslogin/mscii81/dataset/Decision_Tree_ Dataset.csv')
data
```

```python
x=data[['Initial_Payment','Last_Payment','Credit_Score']]
y=data['Result']
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=50)
```

```python
print(X_train)
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
plt.figure(figsize=(12,8))
```

```python
tree.plot_tree(clf,filled=True,feature_names=['Initial_Payment','Last_Payment','Credit_Score'],class_names=['No','Yes'],rounded=True)
plt.show()
```
